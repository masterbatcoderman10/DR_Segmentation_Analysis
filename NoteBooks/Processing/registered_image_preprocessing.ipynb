```python
import os
import shutil
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
base_dir = "../../../Data/retinal-lesions-v20191227/fundus_images/"
annotations_dir = "../../../Data/retinal-lesions-v20191227/lesion_annotations_color_darker/"
target_dir = "../../../Data/retinal-lesions-v20191227/fundus_images_registered/"
```

```python
try:
    os.mkdir(target_dir)
except:
    print("diretory already exists")
```

```python
def image_overlay(img, ann):
        
        """This function overlays the annotation over the image
        This function assumes that the background color is pixel value (0,0,0) in RGB
        img : numpy array in the shape : (height, width, color, alpha)
        ann : numpy array in the shape : (height, width, color, alpha)"""
        
        c = np.where(ann[:,:,[0,1,2]] == [0,0,0])
        c_2 = np.where(ann[:,:,[0,1,2]] != [0,0,0])
        ann[c[0], c[1],-1] = 0
        ann[c_2[0], c_2[1],-1] = 100
        
        img = Image.fromarray(img)
        ann = Image.fromarray(ann)
        
        #Overlaying the annotation on the image
        img.paste(ann, (0,0), ann)
        img = np.array(img)
        
        return img
```

```python
image_files = sorted(list(os.listdir(base_dir)))
image_files =  sorted([os.path.join(base_dir, img_file) for img_file in image_files])

ann_files = sorted(list(os.listdir(annotations_dir)))
ann_files = sorted([os.path.join(annotations_dir, img_file) for img_file in ann_files])
```

```python
counter = 0

for img_file, ann_file in zip(image_files, ann_files):

    file_name = img_file.split("/")[-1]

    img = cv2.imread(img_file)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2RGBA)

    ann = cv2.imread(ann_file)
    ann = cv2.cvtColor(ann, cv2.COLOR_RGB2RGBA)

    reg_img = image_overlay(img, ann)
    reg_img = cv2.cvtColor(reg_img, cv2.COLOR_RGBA2RGB)

    dst_file = os.path.join(target_dir, file_name)
    cv2.imwrite(dst_file, reg_img)

    counter += 1
    print(f"Processed image {counter}")

    
```

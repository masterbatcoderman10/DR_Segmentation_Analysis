```python
import pandas as pd
import numpy as np
import os
```

```python
grades = pd.read_csv("../../../Data/retinal-lesions-v20191227/dr_grades_segs.csv", index_col=0)
```

```python
grades.head()
```

```python
imp_data = grades[["image id", "our label"]]
lesion_data = grades[["image id", "our label", "retinal_hemorrhage", "cotton_wool_spots", "hard_exudate", "microaneurysm"]]
```

```python
data_np = imp_data.to_numpy()
```

```python
np.unique(data_np[:,1])
```

```python
target_dir = "../../../Data/retinal-lesions-v20191227/dr_grades/"
target_dir_plus = "../../../Data/retinal-lesions-v20191227/dr_grades_plus/"
```

```python
tfile = "../../../Data/retinal-lesions-v20191227/dr_grades/9_left.txt"
```

```python
with open(tfile, 'r') as tile:
    i = tile.read()
print(int(i))
```

```python
count = 0

for img in data_np:
    
    file_path = os.path.join(target_dir, (img[0]+".txt"))

    with open(file_path, 'w') as grade_file:
        grade_file.write(str(img[1]))

    count += 1
    print(f"Processed image file {count}")
    
```

```python
lesion_data_np = lesion_data.to_numpy()
```

```python
lesion_data_np.shape
```

```python
count = 0

for img in lesion_data_np:
    
    file_path = os.path.join(target_dir_plus, (img[0]+".txt"))
    with open(file_path, 'w') as grade_file:
        for i in range(1, 6):
            grade_file.write(str(int(img[i])))
            grade_file.write(" ")
    count += 1
    print(f"Finished processing file {count}")

```

```python
import tensorflow as tf
import cv2
from tensorflow.keras.utils import save_img
import os
import shutil
import re
import numpy as np
```

```python
annotations_dir = "../../../Data/retinal-lesions-v20191227/lesion_segs_896x896/"
proper_files = ["cotton_wool_spots.png", "hard_exudate.png", "retinal_hemorrhage.png", "microaneurysm.png"]
file_dict = dict(zip(proper_files, range(4)))
cleaned_dir = "../../../Data/retinal-lesions-v20191227/cleaned_annotations"
try:
    os.mkdir(cleaned_dir)
except OSError:
    print("Directory exists")
pattern = r"\d+_\w+"
```

```python
i = 0

for dp, di, fn in os.walk(annotations_dir):
    for f in fn:
        try:
            match = file_dict[f]
            ann_path = os.path.join(dp, f)
            dir_name = re.findall(pattern, dp)[0]
            
            ann = cv2.imread(ann_path, 0)
            ann = np.clip(ann, None, a_max=1)

            dest_path = os.path.join(cleaned_dir, dir_name, f)

            try:
                cv2.imwrite(dest_path, ann)
            except:
                os.mkdir(os.path.join(cleaned_dir, dir_name))
                cv2.imwrite(dest_path, ann)

        except:

            continue
        
    
    
```

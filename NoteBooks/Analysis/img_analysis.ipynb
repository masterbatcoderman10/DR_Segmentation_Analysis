```python
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
from tensorflow.keras.utils import array_to_img
import os
import shutil 
import pathlib
```

## 1. IDRID Segmentation files

The IDRID dataset contains segmentation maps which have red annotations, this will be explored. Furthermore, these images are in tif format.



```python
idr = "../Data/IDRID/A. Segmentation/2. All Segmentation Groundtruths/a. Training Set/2. Haemorrhages"
```

```python
fst = "IDRiD_17_HE.tif"
```

```python
fst
img_path = os.path.join(idr, fst)
```

```python
img = load_img(img_path)
img_p = Image.open(img_path)
```

```python
img_p
```

```python
img_p.size
```

```python
img.size
```

```python
img_arr = img_to_array(img)
```

```python
img_arr.shape
```

```python
w = img_arr.shape[1]
h = img_arr.shape[0]
b = False

for py in range(h):
    for px in range(w):
        pix = img_arr[py, px]
        if not np.allclose(pix, np.zeros(3)):
            print(pix)
            b = True
            break
    if b: break
```

This result confirms our assumption, the annotations are marked with a red color.
